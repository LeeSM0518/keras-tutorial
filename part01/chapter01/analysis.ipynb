```
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
print(len(x_train))
print(x_train[0])
```

```
print(len(y_train))
print(y_train[0])
```

```
x_train = x_train.reshape(60000, 784).astype('float32') / 255.0
x_test = x_test.reshape(10000, 784).astype('float32') / 255.0
```

```
print(y_train[0])
```

```
y_train = np_utils.to_categorical(y_train)  # 원핫 인코딩
print(y_train[0])
```

```
y_test = np_utils.to_categorical(y_test)
```

```
model = Sequential()
```

```
# 은닉층 구성
model.add(Dense(units=64, input_dim=28*28, activation='relu')) 
model.add(Dense(units=10, activation='softmax'))
```

```
# loss: 현재 가중치를 평가하는데 사용하는 손실함수. categorical_crossentropy 다중 분류 손실함수
# optimizer: 최적의 가중치를 검색하는데 사용되는 최적화 알고리즘. sgd는 우리가 알고 있던 
#   Gradient Descent 보다 좋은 최적화 알고리즘이다.
# metrics : 평가 척도를 나타내며 분류 문제에서는 일반적으로 'accuracy' 으로 지정한다.
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```
# 4. 모델 학습시키기
# 첫 번째 인자 : 입력 변수, X
# 두 번째 인자 : 출력 변수, Y
# epochs : 전체 훈련 데이터셋에 대해 학습 반복 횟수
# batch_size : 가중치를 업데이트할 배치 크기를 의미
hist = model.fit(x_train, y_train, epochs=5, batch_size=32)
```

```
print(hist.history['loss'])       # 손실값
print(hist.history['accuracy'])   # 정확도
```

```
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
```

```
print(loss_and_metrics)
```
